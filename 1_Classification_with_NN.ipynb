```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import re
from tqdm import tqdm 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sn
```

Link to used dataset: https://www.kaggle.com/datasets/team-ai/spam-text-message-classification

```python
texts = pd.read_csv('Spam_text.csv') 
texts
```

##Text class to numeric

```python
texts['Category'] = pd.factorize(texts['Category'])[0]
texts
```

```python
plt.hist(texts['Category'], 
         bins=2)
```

Despite this dataset is very imbalanced, we can exctract enough data to train classification model 

##Text preprocessing


```python
from stop_words import get_stop_words
# https://github.com/Alir3z4/python-stop-words
stop_words = get_stop_words('en')
print(stop_words[:10], len(stop_words))
```

```python
all_words = [] # list for all words in all texts
for row in texts.iterrows():

  text = row[1]['Message']
  clear = re.findall('[a-z]+', text.lower()) # all words in one text

  for word in clear:
    if word in stop_words or len(word) == 1:
      clear.remove(word)

  all_words.extend(clear)
      
  clear = ' '.join(clear)
  texts.loc[row[0], 'Message'] = clear
```

```python
from collections import Counter
word_count = Counter(all_words)
word_count.most_common(20)

all_words = [i[0] for i in word_count.most_common(300)]
```

```python
unique_words = list(set(all_words))
len(unique_words)
```

It is better to perform some analysis to decide how many tokens we should use in one hot. 300 words was chosen based on some experimentation but this number is not the best. <br><br>Also, we could extract top words from each class independently and after put them in one list. This would also improve final result

```python
texts
```

Info about one hot: https://en.wikipedia.org/wiki/One-hot

```python
one_hot_list = []
# custom one hot encoder
for row in tqdm(texts.iterrows()):
  text = row[1]['Message']
  one_hot_text = [1 if i in text else 0 for i in unique_words]
  one_hot_list.append(one_hot_text)

one_hot_list = np.array(one_hot_list)
```

```python
df = pd.DataFrame(one_hot_list, columns = unique_words)
df.insert(loc=0, column='Category', value=texts['Category'])
df
```

##Data preparation

```python
labels = np.array(df['Category'])
data = np.array(df.drop('Category', axis=1))
```

```python
data_train, data_test, label_train, label_test = train_test_split(data, labels, test_size=0.4, random_state=42)
data_train.shape[0], data_test.shape[0]
```

Usually only 20% of dataset is used as test part. But, as I mentioned earlier, this dataset is very convenient, that's why we can increase test

##Model creation

image.png

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(len(unique_words)))
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(64, activation = 'relu'))
model.add(tf.keras.layers.Dense(2, activation = 'softmax'))
model.compile(loss='sparse_categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

##Training

```python
model.fit(data_train, label_train, epochs=10, batch_size=32)
```

##Evaluation

```python
y_pred = model.predict(data_test)
y_pred[0]
```

```python
prediction = np.argmax(y_pred, axis=1)
```

```python
print(classification_report(label_test, prediction))
```

```python
sn.heatmap(confusion_matrix(label_test, prediction), annot=True)
```

This result is enough for demonstration, but it could be significanly improved
